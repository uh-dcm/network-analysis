# Loading a Slack conversation

* This is a fake file similar to what one might get from Slack
* The file is normal JSON file and each has messages
* Each message has user, text, reactions etc.

```python
data = json.load(open('data/fake-slack.json'))
messages = data['messages']
```

```python
messages[0]
```

# Generate a mention network

We check who mentions whom from the dataset

1. Initialise an empty graph
1. Go through all messages and identify who is mentioned
1. Add edges if they have been mentioned

```python
import networkx as nx

G = nx.DiGraph()
```

```python
for message in messages:
    user = message['user']

    if not G.has_node(user): ## add users only if they do not already exist
        G.add_node(user)

    for mention in message['mentions']:
        if not G.has_node(mention): ## check if mention node exists
            G.add_node(mention)

        if not G.has_edge(user, mention): 
            G.add_edge(user, mention, weight=1) ## add an edge
        else:
            G[user][mention]['weight'] += 1 ## increase egde weight
```

```python
G.nodes()
```
